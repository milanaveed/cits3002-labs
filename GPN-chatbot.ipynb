# Part 2-4

```python
import time
print("Welcome to the secret diary chatbot!")
time.sleep(0.5)
name = input("What is your name? ")

print("Hello, " + name)
time.sleep(0.5)

action = input("""What would you like to do?
               1) Write into my diary
               2) Read my diary
               3) Exit
               """)

if action == '1':
    print("You want to write in the diary")
    diary_entry = input("what is your message? ")
    diary = open("secret_diary.txt", "a") # This tells Python to open a file named "secret_diary.txt". If the file doesn’t exist, Python will create it. The "a" means append, which allows you to add new content to the file without deleting the existing text.
    diary_entry = diary_entry + "\n"
    diary.write(diary_entry)
    diary.close()

elif action == '2':
    print("You want to read the diary")
    diary = open("secret_diary.txt", "r")
    content = diary.read()
    print(content)
    diary.close()
    
else:
    print("Goodbye!")
    exit()
```

# Extension 10

```python
import time

password = "open_sesame"

print("Welcome to the secret diary chatbot!")
time.sleep(0.5)

name = input("What is your name? ")
print("Hello, " + name)

guessed_password = input("What's the password? ")
while guessed_password != password:
    guessed_password = input("What's the secret password?")


time.sleep(0.5)

action = input("""What would you like to do?
            1) Write into my diary
            2) Read my diary
            3) Exit
            """)

if action == '1':
    print("You want to write in the diary")
    diary_entry = input("what is your message? ")
    diary = open("secret_diary.txt", "a", encoding="utf-8") # This tells Python to open a file named "secret_diary.txt". If the file doesn’t exist, Python will create it. The "a" means append, which allows you to add new content to the file without deleting the existing text.
    diary_entry = diary_entry.replace("a", "🍏")
    diary_entry = diary_entry.replace("b", "🍌")
    diary_entry = diary_entry.replace("c", "🐈")
    diary_entry = diary_entry.replace("d", "🐶")
    diary_entry = diary_entry.replace("e", "🐘")

    diary_entry = diary_entry + "\n"
    diary.write(diary_entry)
    diary.close()

elif action == '2':
    print("You want to read the diary")
    diary = open("secret_diary.txt", "r")
    content = diary.read()
    content = content.replace("🍏", "a")
    content = content.replace("🍌", "b")
    content = content.replace("🐈", "c")
    content = content.replace("🐶", "d")
    content = content.replace("🐘", "e")
    print(content)
    diary.close()
    
else:
    print("Goodbye!")
    exit()
```
